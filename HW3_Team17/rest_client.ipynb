# Rest API

### Create RestApi Client

```
import requests as r
import matplotlib.pyplot as plt

# Base url of the API server
base_url = 'https://82f6f9dd-e2eb-4faf-a867-1a0a5dea3d6b.deepnoteproject.com'

# Start and end timestamp in milliseconds
start = 0
end = 2000000000

# Get all devices
response = r.get(f'{base_url}/devices')

addresses = response.json()

# For each device get battery info
for mac_address in addresses['mac_addresses']:
    print(mac_address)
    battery_info = r.get(f'{base_url}/device/{mac_address}?from={start}&to={end}')
    battery_info = battery_info.json()
    plt.plot(list(range(len(battery_info['battery_levels']))), battery_info['battery_levels'])
    plt.title(mac_address)
    plt.show()

# Delete last device of the list
response = r.delete(f'{base_url}/device/{addresses["mac_addresses"][-1]}')
if response.status_code == 200:
    print('Device eliminated successfully')
```
